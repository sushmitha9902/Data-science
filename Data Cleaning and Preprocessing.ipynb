# Task 6: Handle Outliers

```python
Q1 = df['Income'].quantile(0.25)  
Q3 = df['Income'].quantile(0.75)  
IQR = Q3 - Q1     
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
df_filtered = df[(df['Income'] >= lower_bound) & (df['Income'] <=upper_bound)]
df_filtered
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age', y='Income', data=df_filtered, palette='coolwarm')
plt.title('Boxplot Before and After Outlier Removal (Seaborn)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
