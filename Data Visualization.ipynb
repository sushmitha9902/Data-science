# 1: Line Plot:

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Year': [2017, 2018, 2019, 2020, 2021],
    'Profit': [80, 120, 100, 150, 160]
})
df
```

```python
plt.figure(figsize=(4,3))
df['Year'] = df['Year'].astype(str)
plt.plot(df['Year'], df['Profit'])
plt.title('Yearly Profit')
plt.xlabel('Year')
plt.ylabel('Profit(in lakhs)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df = pd.DataFrame({
    'Department': ['CS', 'IT', 'ECE', 'ME'],
    'Students': [50, 40, 60, 30]
})
df
```

```python
plt.figure(figsize=(3,4))

plt.bar(df['Department'], df['Students'], color='pink')
plt.title('Students in Department')
plt.xlabel('Department')
plt.ylabel('No of students')
plt.show()
```

```python
df = pd.DataFrame({
    'Sport': ['Cricket', 'Football', 'Basketball', 'Others'],
    'Students': [45, 30, 15, 10]
})
```

```python
df
```

```python
sizes = df['Students']
labels = df['Sport']

plt.pie(sizes, labels=labels, autopct='%1.0f%%')
plt.title('Students in sports')


plt.legend(labels, title='Sport', loc='center')

plt.show()
```

```python
df =pd.DataFrame({
    'Scores': [35, 40, 42, 50, 55, 55, 60, 65, 70, 75, 80, 85, 90]
})
```

```python
df
```

```python
plt.figure(figsize=(4,5))
plt.hist(df['Scores'], bins=5, color='yellow', edgecolor='orange')

plt.title('Score Distribution')
plt.xlabel('Score Range')
plt.ylabel('frequency')

plt.show()
```

```python
df =pd.DataFrame({
    'Height_cm': [150, 155, 160, 165, 170],
    'Weight_kg': [45, 50, 55, 60, 65]
})
```

```python
df
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='smoker', s=120)

plt.title('Scatter Plot of Tips')
plt.xlabel('Total Bill')
plt.ylabel('Tip')

plt.show()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(data=tips, x='day', y='tip',errorbar=None, palette='pastel', hue='day')

plt.title('Average Tip by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Average Tip')

plt.show()
```

```python
sns.histplot(data=tips, x='total_bill', bins=15, kde=True, color='lavender')


plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')


plt.show()
```

```python

flights = sns.load_dataset('flights')
flights.head()
```

```python
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')
flights_pivot
```

```python
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap='Spectral')

# Add title
plt.title('Flights Heatmap')

# Display the plot
plt.show()
```
